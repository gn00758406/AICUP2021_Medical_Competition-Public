```python
import json
import re
from opencc import OpenCC
cc = OpenCC('t2s')
with open("data/Train_qa_ans.json",encoding = 'utf-8') as f:
    data = json.load(f)
```

```python

out_all=[]
for x in range (len(data)):
    sent  = data[x]['text']
    count = sent.count('個管師：')+sent.count('民眾：')+sent.count('家屬：')+sent.count('醫師：')+sent.count('護理師：')
    out = []
    
    for i in range(count):
        try:
            res = re.search(r'(個管師\w*\s*：|民眾\w*\s*：|家屬\w*\s*：|醫師\w*\s*：|護理師\w*\s*：)', sent)
            out1 = sent[:res.end()]
            sent = sent[res.end():]

            res  = re.search(r'(個管師\w*\s*：|民眾\w*\s*：|家屬\w*\s*：|醫師\w*\s*：|護理師\w*\s*：)', sent)

            out2 = sent[:res.start()]
            sent = sent[res.start():]
            out.append([out1+out2])
        except:
            out.append([out1+sent])
            continue
    out_all.append(out)
        
    if len(out_all[x]) != count:
        print("error")
```

```python
stopwords=[]
stopwords.extend(['個','管','師','民','眾','醫','師','家','屬','護','理','師','：','，','的','齁','子','？','。'])
def tokenizer(s):
    words = []
    for word in s:
        if word not in stopwords:
            words.append(word)
    str1 = ''.join(words)
    return str1
```

```python
import heapq
out_text_list=[]
for data_num in range(0,len(data)):   
    out_set=[]
    count_list1=[]
    count_list2=[]
    count_list3=[]

    for g in range(len(out_all[data_num])):
        
        count = 0
        for token_i in out_all[data_num][g][0]:
            for token_o in data[data_num]['question']['choices'][0]['text']:
                if token_o==token_i:
                    count=count+1
                    break
        count_list1.append(count)
        if data[data_num]['question']['choices'][0]['text'] in out_all[data_num][g][0]:
            out_set.append(out_all[data_num][g][0])

    sort = list(heapq.nlargest(2, enumerate(count_list1), key=lambda x:x[1]))
    if out_all[data_num][sort[0][0]][0] not in out_set:
        out_set.append(out_all[data_num][sort[0][0]][0])
        

        if out_all[data_num][sort[0][0]][0][-1]=='？' or out_all[data_num][sort[0][0]][0][-1]=='?':
            try:
                out_set.append(out_all[data_num][sort[0][0]+1][0])
            except:
                print("error")

    if out_all[data_num][sort[1][0]][0] not in out_set:
        out_set.append(out_all[data_num][sort[1][0]][0])

        if out_all[data_num][sort[1][0]][0][-1]=='？' or out_all[data_num][sort[1][0]][0][-1]=='?':
            try:
                out_set.append(out_all[data_num][sort[1][0]+1][0])
            except:
                print("error")



    for g in range(len(out_all[data_num])):
        
        count = 0
        for token_i in out_all[data_num][g][0]:
            for token_o in data[data_num]['question']['choices'][1]['text']:
                if token_o==token_i:
                    count=count+1
                    break
        count_list2.append(count)
        if data[data_num]['question']['choices'][1]['text'] in out_all[data_num][g][0]:
            out_set.append(out_all[data_num][g][0])

    sort2 = list(heapq.nlargest(2, enumerate(count_list2), key=lambda x:x[1]))
    if out_all[data_num][sort2[0][0]][0] not in out_set:
        out_set.append(out_all[data_num][sort2[0][0]][0])
        print(out_all[data_num][sort2[0][0]][0])
        if out_all[data_num][sort2[0][0]][0][-1]=='？' or out_all[data_num][sort2[0][0]][0][-1]=='?':
            try:
                out_set.append(out_all[data_num][sort2[0][0]+1][0])
            except:
                print("error")
    if out_all[data_num][sort2[1][0]][0] not in out_set:
        out_set.append(out_all[data_num][sort2[1][0]][0])

        if out_all[data_num][sort2[1][0]][0][-1]=='？' or out_all[data_num][sort2[1][0]][0][-1]=='?':
            try:
                out_set.append(out_all[data_num][sort2[1][0]+1][0])
            except:
                print("error")


    for g in range(len(out_all[data_num])):
        
        count = 0
        for token_i in out_all[data_num][g][0]:
            for token_o in data[data_num]['question']['choices'][2]['text']:
                if token_o==token_i:
                    count=count+1
                    break
        count_list3.append(count)
        if data[data_num]['question']['choices'][2]['text'] in out_all[data_num][g][0]:
            out_set.append(out_all[data_num][g][0])

    sort3 = list(heapq.nlargest(2, enumerate(count_list3), key=lambda x:x[1]))
    if out_all[data_num][sort3[0][0]][0] not in out_set:
        out_set.append(out_all[data_num][sort3[0][0]][0])

        if out_all[data_num][sort3[0][0]][0][-1]=='？' or out_all[data_num][sort3[0][0]][0][-1]=='?':
            try:
                out_set.append(out_all[data_num][sort3[0][0]+1][0])
            except:
                print("error")

    if out_all[data_num][sort3[1][0]][0] not in out_set:###去除重複
        out_set.append(out_all[data_num][sort3[1][0]][0])

        if out_all[data_num][sort3[1][0]][0][-1]=='？' or out_all[data_num][sort3[1][0]][0][-1]=='?':
            try:
                out_set.append(out_all[data_num][sort3[1][0]+1][0])
            except:
                print("error")

    out_text=""
    for ele in out_set:
        out_text+="".join(ele)

    out_text_list.append(out_text)
```

```python
#將判斷何者為錯誤的問題改成判斷何者為正確，並將正確答案的label做改變
question_list=[]
tmp_list=[]
for x in range(len(data)):
    
    tmp=data[x]['question']['stem']
    
    if "有誤" in tmp:
        tmp=tmp.replace("有誤","正確")
        tmp_list.append(x)
        
    if "不是" in tmp:
        tmp=tmp.replace("不是","是")
        tmp_list.append(x)
        
    if "錯誤" in tmp:
        tmp=tmp.replace("錯誤","正確")
        tmp_list.append(x)
  
    if "不正確" in tmp:
        tmp=tmp.replace("不正確","正確")
        tmp_list.append(x)
    if "不" in tmp:
        tmp=tmp.replace("不","有")
        tmp_list.append(x)


    question_list.append(tmp)
```
